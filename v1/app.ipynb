```python
import json
from selenium import webdriver
from selenium_stealth import stealth
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time
```

```python
options = Options()
# service = Service(executable_path="chrome_driver/chromedriver.exe")
options.add_argument("--disable-blink-features=AutomationControlled")
# driver = webdriver.Chrome(service=service, options=options)
driver = webdriver.Chrome(options=options)
stealth(driver,
        languages=["en-US", "en"],
        vendor="Google Inc.",
        platform="Win32",
        webgl_vendor="Intel Inc.",
        renderer="Intel Iris OpenGL Engine",
        fix_hairline=True,
)
```

```python
from scripts import load_cookies

# opening the site first to set domain
driver.get("https://www.linkedin.com")  

# Load cookies from JSON file
with open("cookies.json", "r") as f:
    cookies = json.load(f)    
load_cookies(driver)
```

```python
from scripts import search_jobs

search_jobs(driver, keyword="Data Scientist", location="New York, United States")
print("Job Found")
```
